```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['team_1','team_2','city'])
]
,remainder='passthrough')
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
pipe1 = Pipeline(steps=[
    ('step1',trf),
    ('step3',SVC(kernel = 'linear', random_state = 0))
])
pipe1.fit(X_train,y_train)
y_pred = pipe1.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
#classifier = 
pipe2 = Pipeline(steps=[
    ('step1',trf),
    ('step2',GaussianNB())
])
pipe2.fit(X_train,y_train)
y_pred = pipe2.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier

pipe3 = Pipeline(steps=[
    ('step1',trf),
    ('step2',DecisionTreeClassifier(criterion = 'entropy', random_state = 0))
])
pipe3.fit(X_train,y_train)
y_pred = pipe3.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
#classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
pipe4 = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0))
])
pipe4.fit(X_train,y_train)
y_pred = pipe4.predict(X_test)
accuracy_score(y_test,y_pred)
```
